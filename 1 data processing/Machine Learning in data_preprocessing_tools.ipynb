# Data Preprocessing Tools

## Importing the libraries

```
 import numpy as np
 import matplotlib.pyplot as plt
 import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
dataset
```

```
x = dataset.iloc[:, :-1].values   # we taking all rows and all col except last col
y = dataset.iloc[:,-1].values   # only last col
```

```
print(x)
```

```
print(y)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer    ## this is class as oops
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')  ## we replacing with mean value
imputer.fit(x[:,1:3])
x[:,1:3] = imputer.transform(x[:,1:3]) # chnge in 1 and 2 column
```

```
print(x)     # now we get the filled matrix in place of nan
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder',OneHotEncoder(),[0])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```
print(x)  # it transform country in code 
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print(y)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 1)
```

```
print(x_train)             ## taking in training the 80% of dataset
```

```
print(x_test)              ## 20 percent of dataset
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

it should be after the tarining and test spliting to avoid information leak

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:,3:] = sc.fit_transform(x_train[:,3:])
x_test[:,3:] = sc.transform(x_test[:,3:])
```

```
print(x_train)
```
