solve an ordinary least squares regression problem

```python
n_train = len(data_train.index)
print(n_train)
n_test = len(data_test)
print(n_test)
```

```python
cols = list(np.arange(1,18))
x_train = data_train[data_train.columns[cols]]
x_test = data_test[data_test.columns[cols]]
x_train.head()
```

```python
y_train = data_train[data_train.columns[0]].to_numpy()
y_test = data_test[data_test.columns[0]].to_numpy()

y_train = y_train.reshape(357,1)
y_test = y_test.reshape(89,1)
```

```python
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
x_test.shape , x_train.shape
```

```python
x_train
```

```python
# adding column of 1 at last

x_train = np.hstack((x_train, np.ones((x_train.shape[0],1),dtype=x_train.dtype)))
x_test = np.hstack((x_test, np.ones((x_test.shape[0],1),dtype=x_test.dtype)))
```

```python
# solving least square

XTX = np.matmul(np.transpose(x_train),x_train)

print('XTX',XTX.shape)

XTY = np.matmul(np.transpose(x_train),y_train)

print('XTY' , XTY.shape)

beta = np.linalg.solve(XTX,XTY)

print('regression coefficient', beta)
```

R^2 value and mean sqaure based on training data

```python
# calculating y_bar

y_bar = 0

for i in range(n_train):
    y_i = float(data_train.iat[i,0])
    y_bar += y_i
y_bar /= n_train

print(y_bar)
print(np.mean(y_train))
```

```python
# computing sigma_yy 

sigma_yy = 0

for i in range(n_train):
    y_i_train = float(data_train.iat[i,0])
    sigma_yy += (y_i_train - y_bar)**2

print('sigma_yy', sigma_yy)
```

```python
sum_sq_residuals = 0

for i in range(n_train):
    x_i = x_train[i,:]
    y_i = y_train[i]

    y_pred_i = np.dot(x_i,beta)
    e_i = y_i - y_pred_i

    sum_sq_residuals += (e_i)**2

print('sum of squared residuals ', sum_sq_residuals)

R_SQ = 1-(sum_sq_residuals/sigma_yy)
print('R_SQ', R_SQ)

MSE_TRAIN = sum_sq_residuals/n_train

print('MSE_TRAIN', MSE_TRAIN)
```

R^2 value and mean sqaure based on testing data

```python
y_bar_test = 0
for i in range(n_test):
    y_i_test = float(data_test.iat[i,0])
    y_bar_test += y_i_test
y_bar_test /= n_test

print(y_bar_test)
# or
print(np.mean(y_test))
```

```python
sigma_yy_test = 0



for i in range(n_test):
    y_i_test = float(data_test.iat[i,0])
    sigma_yy_test += (y_i_test - y_bar_test)**2

print('sigma_yy', sigma_yy_test)
```

```python
sum_sq_residuals_test = 0

for i in range(n_test):
    x_i_test = x_test[i,:]
    y_i_test = y_test[i]

    y_pred_i = np.dot(x_i_test, beta)

    e_i = y_i_test - y_pred_i

    sum_sq_residuals += (e_i)**2

print('sum of sqaured error on test data', sum_sq_residuals_test)

R_sq_test = 1-(sum_sq_residuals_test/sigma_yy_test)

print('R_sq_test', R_sq_test)

MSE_TEST = sum_sq_residuals_test/n_test

print('MSE_TEST',MSE_TEST)
```

```python
seed = 150 

data_train_initial = data.sample(frac=0.8,random_state=seed)
data_test = data.drop(data_train_initial.index)
```

```python
print(data_train_initial.shape)
print(data_test.shape)
```

defining the function for the R2 and MSE values

```python
def Rq_and_MSE(data, beta, X, Y): 

  
  y_bar = 0.
  n = len(data)
  for i in range(n): 
    y_i= float(data.iat[i,0]) 
    y_bar += y_i

  y_bar /= n
  

  sigma_yy = 0
  for i in range(n): 
    y_i = float(data.iat[i,0]) 
    sigma_yy += (y_i-y_bar)**2

  
  sum_sq_residuals = 0.
  for i in range(n):  
    x_i = X[i,:] 
    y_i = Y[i] 

    y_pred_i = np.dot(x_i,beta) 
    e_i = y_i - y_pred_i 

    sum_sq_residuals += (e_i)**2

    
  
  R_sq = 1-sum_sq_residuals/sigma_yy
  MSE = sum_sq_residuals/n
  print('R^2 :',R_sq, ' MSE:',MSE)
  return R_sq, MSE
```

```python
seeds = [200,250,300,350,400]

train_R2_seeds = []
train_MSE_seed = []

colsx = list(np.arange(1,18))
res = [0]

for seed in seeds:
    data_train_final = data_train_initial.sample(frac=0.8 , random_state=seed)

    x_train = data_train_final[data_train_final.columns[colsx]].to_numpy()
    y_train = data_train_final[data_train_final.columns[res]].to_numpy()

    # adding ones col

    x_train = np.hstack((x_train,np.ones((x_train.shape[0],1),dtype=x_train.dtype)))
    
    train_R2_lambdas = []
    train_MSE_lambdas = []

    for lambda_ in lambdas:
        XTX = np.matmul(np.transpose(x_train),x_train)
        XTY = np.matmul(np.transpose(x_train),y_train)
        I = np.identity(x_train.shape[1])

        beta = scipy.linalg.solve(XTX + lambda_*I , XTY)
        
        print('lambda', lambda_)
        print('beta', beta)
        print(">>>>>>>>>>>>>>>>>>>>>>>>>")

        train_r2, train_mse = Rq_and_MSE(data_train_final,beta,x_train,y_train)
        train_R2_lambdas.append(train_r2)
        train_MSE_lambdas.append(train_mse)

    train_R2_seeds.append(np.array(train_R2_lambdas))
    train_MSE_seed.append(np.array(train_MSE_lambdas))

    print(">>>>>>>>>>>>>>>>>>>************")
    print("for next seed", seed)

    

```
